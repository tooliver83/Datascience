```python
df = pd.read_csv("data/API_SP.DYN.LE00.IN_DS2_en_csv_v2_10181296.csv", header=2, index_col=(0,1,2,3))
df.head()
```

- Head function returns top 5 rows
- By indexing columns, the index columns and the numeric columns look different (numeric columns are higher)

```python
life_expectancy = df.groupby(level=0).first().transpose()
life_expectancy.head()
```

- Clean up the data by removing NaN - first by columns, and then by rows

```python
life_expectancy_clean = life_expectancy.dropna(axis=1, thresh=45)
life_expectancy_clean.dropna(how='all')
```

- Add 'mean' row at the bottom the data to sort largest countries

```python
df.loc['mean'] = life_expectancy_clean.mean()
```

```python
subset = life_expectancy_clean.sort_values(life_expectancy_clean.last_valid_index(),axis=1)
subset.head()
```

- Plot the top 5 and bottom 5 countries (separate graphs) by mean life expectancy.
- Do this by taking our sorted dataset, then finding number of columns, and using this to find the high dataset.
- Remove mean and 2017 rows (either NaN or misleading)

```python
x = len(subset.columns)
subset_low = subset.iloc[:-3, 0:5]
subset_low.plot()

subset_high = subset.iloc[:-3, x-5:x]
subset_high.plot()
```
